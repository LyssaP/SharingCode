```python
#importing my packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing my data into a dataframe
trip_advisor = pd.read_csv(r'tripadvisor_review.csv')
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1,2,1)
plt.plot(trip_advisor['Juice Bars'],color='hotpink')
plt.xlabel('Users',fontsize=18)
plt.ylabel('Scores',fontsize=14)
plt.title('Juice Bar Scores',fontsize=26)

plt.subplot(1,2,2)
plt.plot(trip_advisor['Restaurants'],color='purple')
plt.xlabel('Users',fontsize=18)
plt.ylabel('Scores',fontsize=14)
plt.title('Restaurant Scores',fontsize=26)
```

```python
#Looking at both of these graphs side by side, it is evident that traveler's much preferred the juice bars of East Asia to the
#restaurants. From this, I would recommend either adding more juice bars to areas with lower enjoyment or look into why 
#restaurants do not provide as pleasant an experience.
```

```python
#Made a new column to give me insight into each traveler's overall happiness with their trip to East Asia
trip_advisor['Overall Enjoyment'] = trip_advisor.mean(axis=1)
```

```python
plt.figure(figsize=(20,5))
plt.scatter(trip_advisor['Overall Enjoyment'],y=trip_advisor['Resorts'],color='green')
plt.scatter(trip_advisor['Overall Enjoyment'],y=trip_advisor['Beaches'],color='orange')
plt.scatter(trip_advisor['Overall Enjoyment'],y=trip_advisor['Parks/ Picnics'],color='cyan')
plt.xlabel('Overall Enjoyment',fontsize=18)
plt.ylabel('Scores',fontsize=16)
plt.title('Effect of Resorts, Beaches, and Parks on Overall Enjoyment',fontsize=22)
print("""
Green = Resort
Orange = Beaches
Light Blue = Parks
""")
```

```python
#From this graph's analysis, I would highly recommend focusing on making sure that resorts in East Asia are top notch because
#there is a very evident positive relationship between traveler's enjoyment of resorts and their overall enjoyment of their 
#trip. On the contrary, beaches and parks, especially parks, have little to no influence on overall enjoyment.
```

```python
plt.figure(figsize=(20,5))
plt.hist(trip_advisor['Museums'],bins=100,color='red',alpha=.5)
plt.hist(trip_advisor['Dance Clubs'],bins=100,color='blue',alpha=.5)
plt.title("Comparison of Traveler's Interest in Museums vs Dance Clubs",fontsize=22)
plt.xlabel('Scores',fontsize=18)
plt.ylabel('Number of Travelers',fontsize=16)
print("""
Red = Museums
Blue = Dance Clubs
""")
```

```python
#Both graphs are skewed to the right, telling me that both are below average in likability. That being said, dance clubs are
#closer to center and have more higher scores than museums do, so, assuming that both are visited equally as much, I would 
#recommend focusing on improving museums.
```
